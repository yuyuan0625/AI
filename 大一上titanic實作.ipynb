```
# -*- coding: utf-8 -*-
"""
Created on Tue Dec 17 22:13:29 2019

@author: user
"""

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd

train= pd.read_csv("train.csv")
train= train[['Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']].dropna()

X= train[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
target= pd.DataFrame(train,columns=['Survived'])
y= target["Survived"]

for i in range(len(X)):
    if X.iat[i,1] == 'male':
        X.iat[i,1]=0
    else:
        X.iat[i,1]=1
for i in range(len(X)):
    if X.iat[i,6] == 'C':
        X.iat[i,6]=0
    elif X.iat[i,6] == 'Q':
        X.iat[i,6]=1        
    else:
        X.iat[i,6]=2

train_data,test_data,train_label,test_label= train_test_split(X,y,test_size=0.2)
k=3
knn= KNeighborsClassifier(n_neighbors=k)

knn.fit(train_data,train_label)

print('準確率=',knn.score(test_data,test_label))

from sklearn.metrics import confusion_matrix
pred_y= knn.predict(test_data)
print('預測是否存活(0 = No, 1 = Yes)')
print(pred_y)
print('混淆矩阵=')
print(confusion_matrix(test_label,pred_y))
```
