
### size、torch.squeeze()

```python
x = torch.zeros(2,1,2,1,2) #產生全0的陣列
x.size()
```

```python
x
```

```python
y = torch.squeeze(x) #消掉多餘維度
y.size()
```

```python
y
```

```python
y = torch.squeeze(x,0) #可指定squeeze的index(如果可壓縮)
y.size()
```

```python
y = torch.squeeze(x,1) #可指定squeeze的index(如果可壓縮)
y.size()
```

```python
y = torch.squeeze(y, 2) #注意前面若squeeze後，index會改變
y.size()
```
