```
from scipy import stats
import pandas as pd

def Entropy(labels, base=2):
  #計算機率
  probs = pd.Series(labels).value_counts() / len(labels)
  en = stats.entropy(probs, base=base)
  return en
```

```
#計算一組數字的亂度、數值越高代表亂度越高
print(Entropy([1, 1, 0, 1, 1]))
print(Entropy([1, 1, 0, 1, 0]))
```
