```
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules, fpgrowth
```

```
#read the data
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/retail_dataset.csv')
df
```

```
#find out how many unique items are actually there in the table
items = (df['0'].unique())
items
```

```
#One Hot Encode the data
encoded_vals = []
for index, row in df.iterrows():
  labels = {}
  uncommons = list(set(items) - set(row))
  commons = list(set(items).intersection(row))
  for uc in uncommons:
    labels[uc] = 0
  for com in commons:
    labels[com] = 1
    encoded_vals.append(labels)
encoded_vals[0]

ohe_df = pd.DataFrame(encoded_vals)
ohe_df
```

```
#Applying Apriori
freq_items = apriori(ohe_df, min_support=0.2, use_colnames=True)
freq_items.head(7)
```

```
#adjust the metric and min_threshold srguments
#eg.if you want to see rules which have a lift score of >=1.0
rules = association_rules(freq_items, metric="lift", min_threshold = 1.0)
rules.head()
```

```
#FProwth
fpgrowth.fpgrowth(ohe_df, min_support=0.2, use_coluname=True)
```
